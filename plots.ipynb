```python
import plotly.express as px
import numpy as np
from dataloader import df
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
print("\n".join(df.columns))
```

```python
df = df.sort_values(by="timestamp")
```

```python
subplt = make_subplots(rows=3, cols=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["T1_remote_K"].values, mode='lines', name="T1 [K]"), row=1, col=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["T2_embeded_K"].values, mode='lines', name="T2 [K]"), row=1, col=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["DeltaT_K"].values, mode='lines', name="DeltaT [K]"), row=2, col=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["SpDeltaT_K_Write"].values, mode='lines', name="SpDeltaT [K]"), row=2, col=1) 
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["Heating_E_J"].values, mode='lines', name="Heating [J]"), row=3, col=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["Cooling_E_J"].values, mode='lines', name="Cooling [J]"), row=3, col=1)
subplt.add_trace(go.Scatter(x=df["timestamp"], y=df["Cooling_E_J"].values, mode='lines', name="Cooling [J]"), row=3, col=1)


subplt.update_layout(height=1000)
subplt.show()
```
